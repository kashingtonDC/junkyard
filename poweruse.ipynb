```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = [x for x in os.listdir(os.getcwd()) if "ElectricityBy" in x]
```

```python
df = pd.read_csv(file[0])
```

```python
dat = []
dat.append([float(x) for x in df.T.values[2:] ])
```

```python

d = [item for sublist in dat for item in sublist]
```

```python
plt.plot(d[:-1])
```

```python
plt.figure(figsize = (8,3))
plt.plot(df.T.iloc[2:-1][::-1].index, df.T.iloc[2:-1].values[::-1])
plt.title("Tulare County Non Residential Power Use (kWh)")

plt.tight_layout()
plt.axvspan(df.T.index[-18], df.T.index[-21], alpha=0.3, color='red')
plt.axvspan(df.T.index[-23], df.T.index[-27], alpha=0.3, color='red')
plt.axvspan(df.T.index[-10], df.T.index[-15], alpha=0.3, color='red', label = 'drought')
plt.xticks(rotation=90)

plt.legend()
plt.ylabel("kWh")
```
