```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read pivot table
xls = pd.ExcelFile("FVS_4_1.xlsx")

sdfs = {}

for scen in ['lowscenario','defscenario','highscenario']:
    
    df = pd.read_excel(xls, scen)
    # cleanup df
    df.set_index("Year", inplace = True)
    # biomass --> co22
    sdfs[scen] = df * 0.45 * 44/12
    
```

```python
stems_ac = {
    "Base_BK" : 40, # guessed from VT HL
    "Base_HL" : 41, # taken from VT
    "Base_LP" : 50, # made up
    "Base_WN" : 36, # taken from VT
}

for k,v in stems_ac.items():
    
    low = sdfs['lowscenario'][k] * v
    med = sdfs['defscenario'][k] * v
    high = sdfs['highscenario'][k] * v
    
    plt.title(k)
    plt.plot(med, label = "{} {} stems/ac".format(k[-2:],v ), alpha = 0.7)
    plt.fill_between(low.index,low,high, alpha = 0.25)
    plt.legend()
    plt.ylabel("T CO2e / acre")
    plt.ylim(0,25)
    plt.xlim(2022,2060)

    plt.show()
```
