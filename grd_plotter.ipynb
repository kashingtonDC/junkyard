```python
import os
import numpy as np
import rasterio as rio
import matplotlib.pyplot as plt
```

```python
files = [x for x in os.listdir(os.getcwd()) if x.endswith(".grd")]

for file in files:
    outfn = file.split(".")[0] + ".tif"
    warp_cmd = 'gdalwarp {} {}'.format(file, outfn)
    print(warp_cmd)
    os.system(warp_cmd)
    
```

```python
tiffs = [x for x in os.listdir(os.getcwd()) if x.endswith(".tif")]
```

```python
s1_tiff = tiffs[0]
src = rio.open(s1_tiff)

data = src.read(1)

plt.imshow(data)
plt.axis("off")
plt.colorbar()
```
