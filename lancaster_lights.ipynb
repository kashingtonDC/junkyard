```python
import numpy as np
import rasterio as rio
import matplotlib.pyplot as plt
```

```python
src = rio.open("LC09_L1GT_111212_20211125_20220123_02_T2_B8.TIF")
arr = src.read(1)
arr = arr.astype(float)
arr[arr==src.nodata] = np.nan
```

```python
plt.figure(figsize=(10,10)); plt.imshow(arr); plt.title("Band 8"); plt.colorbar(); plt.show()
```
