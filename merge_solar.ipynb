```python
import geopandas as gp
```

```python
gdf_solar = gp.read_file("uspvdbGeoJSON/uspvdb_v2_0_20240801.geojson")
gdf_public = gp.read_file("USA_Federal_Lands_1879006603790792540.geojson")
```

```python
gdf_solar[gdf_solar['p_state'] != "AK"][2:3].plot()
```

```python
gdf_solar_j = gdf_solar.sjoin(gdf_public, how="left")
```

```python
gdf_solar_j['Agency'].dropna().value_counts().plot(kind = 'bar')
```

```python
gdf_solar_j.to_file("sol_util_joined.geojson")
```

```python
gdf_solar_j.to_csv("sol_util_joined.csv")
```
